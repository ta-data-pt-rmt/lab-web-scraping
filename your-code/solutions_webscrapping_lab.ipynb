````python
response = requests.get(url)

# Specify the encoding to prevent from character encoding issues
response.encoding = 'utf-8'

# Parse the HTML
soup = BeautifulSoup(response.text, 'html.parser')

# Find all language boxes
language_boxes = soup.find_all('div', class_='central-featured-lang')

for language_box in language_boxes:
    language_name = language_box.find('a')['title'].split(' — ')[0].strip()
    article_count = language_box.find('bdi').text.replace('\xa0', '').replace('+', '')
    
    print(f'{language_name}: {article_count} articles')
````

````python
response = requests.get(url)

response.encoding = 'utf-8'

# Parse the HTML
soup = BeautifulSoup(response.text, 'html.parser')

# Create empty lists to store data
languages = []
article_counts = []
links = []

# Find all language lists and bookshelf containers
lang_lists = soup.find_all('div', class_='lang-list-content')
bookshelf_containers = soup.find_all('h2', class_='bookshelf-container')

# Extract data from language lists
for lang_list in lang_lists:
    # Extract article count
    article_count = lang_list.find('bdi').text.replace('\xa0', '').replace('+', '')
    
    # Extract language names and links
    lang_items = lang_list.find('div', class_='langlist').find_all('li')
    
    # Iterate through each language item in the list
    for lang_item in lang_items:
        lang_name = lang_item.find('a').text.strip()  # Strip leading/trailing whitespaces
        lang_link = lang_item.find('a')['href']
        
        # Append data to lists
        languages.append(lang_name)
        article_counts.append(article_count.replace('\n', ''))  # Remove newline characters
        links.append(lang_link)

# Extract data from bookshelf containers
for bookshelf_container in bookshelf_containers:
    # Extract article count
    article_count = bookshelf_container.find('bdi').text.replace('\xa0', '').replace('+', '')
    
    # Extract language names and links
    lang_items = bookshelf_container.find_next('div', class_='langlist').find_all('li')
    
    # Iterate through each language item in the list
    for lang_item in lang_items:
        lang_name = lang_item.find('a').text.strip()  # Strip leading/trailing whitespaces
        lang_link = lang_item.find('a')['href']
        
        # Append data to lists
        languages.append(lang_name)
        article_counts.append(article_count.replace('\n', ''))  # Remove newline characters
        links.append(lang_link)

# Create a DataFrame
wiki_df = pd.DataFrame({
    'Language': languages,
    'Article Count': article_counts,
    'Link': links
})

# Set the index starting from 1
wiki_df.index = wiki_df.index + 1

# Print the DataFrame
wiki_df
````

````python
# Understand the page construction to get the info. Example with firts movie :

# got the url
movie_url = 'https://www.imdb.com/title/tt0111161/?ref_=chttp_i_1'

# make the request

response = requests.get(movie_url,headers=headers)

soup = BeautifulSoup(response.content)

soup
````

````python
import json

# Extract JSON-LD script
json_ld_script = soup.find('script', {'type': 'application/ld+json'}).contents[0]

# Load JSON data
movie_data = json.loads(json_ld_script)

# Extract movie name
movie_name = movie_data['name']
print("Movie Name:", movie_name)

# Extract release year
release_year = movie_data['datePublished']
print("Release Year:", release_year)

# Extract director
director_name = movie_data['director'][0]['name']
print("Directed By:", director_name)

# Extract IMDb rating
imdb_rating = movie_data['aggregateRating']['ratingValue']
print("IMDb Rating:", imdb_rating)
````

````python
#Create a function to apply it to all movies 

def get_movie_info(movie_url):
    # Make the request
    response = requests.get(movie_url, headers=headers)
    soup = BeautifulSoup(response.content, 'html.parser')

    # Extract JSON-LD script
    json_ld_script = soup.find('script', {'type': 'application/ld+json'}).contents[0]

    # Load JSON data
    movie_data = json.loads(json_ld_script)

    # Extract movie information
    movie_name = movie_data.get('name', 'N/A')
    release_year = movie_data.get('datePublished', 'N/A')
    director_name = movie_data['director'][0]['name'] if 'director' in movie_data else 'N/A'
    imdb_rating = movie_data['aggregateRating']['ratingValue'] if 'aggregateRating' in movie_data else 'N/A'

    return {
        'Movie Name': movie_name,
        'Release Year': release_year,
        'Directed By': director_name,
        'IMDb Rating': imdb_rating
    }

# Get information for all movies
movie_info_list = []

for movie_url in movie_links[:250]:
    movie_info = get_movie_info(movie_url)
    movie_info_list.append(movie_info)

# Create a DataFrame from the list of movie information
df = pd.DataFrame(movie_info_list)

print(df)
````

````python
if response.status_code == 200:
    # Parse the content
    soup = BeautifulSoup(response.content, 'html.parser')

    # Find the movie entries
    movie_entries = soup.select('.lister-item-content')[:10]

    # Initialize lists to store data
    movie_names = []
    years = []
    summaries = []

    # Loop through the first 10 movie entries
    for movie_entry in movie_entries:
        # Extract movie name
        movie_name = movie_entry.find('a').get_text(strip=True)

        # Extract year
        year = movie_entry.find('span', class_='lister-item-year').get_text(strip=True)

        # Extract summary
        summary = movie_entry.find('p', class_='').get_text(strip=True)

        # Append data to lists
        movie_names.append(movie_name)
        years.append(year)
        summaries.append(summary)

    # Create a DataFrame
    movies_df = pd.DataFrame({
        'Movie Name': movie_names,
        'Year': years,
        'Summary': summaries
    })

    
    print(movies_df)
else:
    print(f"Failed to retrieve the page. Status code: {response.status_code}")
````

#### Find the book name, price and stock availability as a pandas dataframe.

````python
# This is the url you will scrape in this exercise. 
# It is a fictional bookstore created to be scraped. 
url = 'http://books.toscrape.com/'
````

````python
# your code here
response = requests.get(url)

if response.status_code == 200:
    soup = BeautifulSoup(response.content, 'html.parser')

    # Initialize lists to store data
    book_names = []
    prices = []
    availabilities = []

    # Find all product pods
    product_pods = soup.find_all('article', class_='product_pod')

    # Loop through each product pod
    for product_pod in product_pods:
        # Extract book name
        book_name = product_pod.find('h3').find('a')['title']

        # Extract price
        price = product_pod.find('p', class_='price_color').get_text(strip=True)

        # Extract stock availability
        availability = product_pod.find('p', class_='instock availability').get_text(strip=True)

        # Append data to lists
        book_names.append(book_name)
        prices.append(price)
        availabilities.append(availability)

    # Create a DataFrame
    books_df = pd.DataFrame({
        'Book Name': book_names,
        'Price': prices,
        'Availability': availabilities
    })

    # Display the DataFrame
    print(books_df)
else:
    print(f"Failed to retrieve the page. Status code: {response.status_code}")
````

####  Display the 100 latest earthquakes info (date, time, latitude, longitude and region name) by the EMSC as a pandas dataframe.
***Hint:*** Here the displayed number of earthquakes per page is 20, but you can easily move to the next page by looping through the desired number of pages and adding it to the end of the url.

````python
# This is the url you will scrape in this exercise
url = 'https://www.emsc-csem.org/Earthquake/?view='

# This is how you will loop through each page:
number_of_pages = int(100/20)
each_page_urls = []

for n in range(1, number_of_pages+1):
    link = url+str(n)
    each_page_urls.append(link)
    
each_page_urls
````

````python
# List to store earthquake data
data = []

# Loop through each page URL
for page_url in each_page_urls:
    response = requests.get(page_url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        earthquake_rows = soup.find_all('tr', class_='lilist')

        for row in earthquake_rows:
            earthquake_id = row['data-id']
            date_time = row.find('td', class_='tbdat').a.text.strip()
            latitude = float(row.find('td', class_='tblat').text.strip())
            longitude = float(row.find('td', class_='tblon').text.strip())
            region_name = row.find('td', class_='tbreg').text.strip()

            data.append([earthquake_id, date_time, latitude, longitude, region_name])

#Create df :
columns = ['ID', 'Date/Time', 'Latitude', 'Longitude', 'Region']
df = pd.DataFrame(data, columns=columns)


print(df)
````

````python
# Get request

import requests
def geturl(url):
    res = requests.get(url, timeout=15)
    return {'status': res.status_code,
            'content': res.text}
````

````python
#Download the result
res = geturl(url)
print(res['content'])
````

````python
#DF with 100 lastest earthquake as of 29.11.2023 - 2:30pm

import json

content_earthquake = res['content']

# Load the JSON data
earthquake_json = json.loads(content_earthquake)

# Extract the relevant information for the DataFrame
selected_fields = ["time", "lat", "lon", "flynn_region"]
earthquake_data = []

for feature in earthquake_json.get("features", []):
    properties = feature.get("properties", {})
    entry_data = {
        key: properties.get(key, None) 
        for key in selected_fields
    }
    earthquake_data.append(entry_data)

df = pd.DataFrame(earthquake_data)

# Convert the "time" column to a readable datetime format
df["time"] = pd.to_datetime(df["time"])

# Separate "time" into "date" and "time" columns for more readability
df["date"] = df["time"].dt.date
df["time"] = df["time"].dt.time

df = df.rename(columns={"flynn_region": "region"})

df.head(100)
````
